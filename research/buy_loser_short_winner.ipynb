```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_table('/mnt/c/Users/dhu/Documents/quantitative-trading-how-to-build-your-own-algorithmic-trading-business/Python codes/Python codes/SPX_20071123.txt')
```

```python
df['Date'] = df['Date'].astype('int')
```

```python
df.set_index('Date', inplace=True)
df.sort_index()
```

```python
df
```

```python
daily_ret = df.pct_change()
market_daily_ret = daily_ret.mean(axis=1)
```

```python
daily_ret
```

```python
market_daily_ret
```

```python
# calculate return deviation from the mean of the day
weights = np.array(daily_ret) - np.array(market_daily_ret).reshape(daily_ret.shape[0], 1)
weights
```

```python
# weigh the portfolio by daily performance
# each weight is calculated as w_i / sum of abs(w_i). 
# This strategy is dollar neutral because the total dollar in long position = total dollar in short position 
wtsum = np.nansum(abs(weights), axis=1)
wtsum
```

```python
weights[wtsum==0,] = 0
wtsum[wtsum==0] = 1 # hack to make sure the following division works 
```

```python
weights=weights/wtsum.reshape((daily_ret.shape[0], 1))
```

```python
weights
```

```python
weights *= -1
weights # flip the sign for trading direction, better performance, more short exposure
```

```python
# important, shift weights derived from previous way so today's trading uses previous signal (return)
daily_pnl = np.nansum(np.array(pd.DataFrame(weights).shift()) * np.array(daily_ret), axis=1)
daily_pnl
```

```python
daily_pnl_test = daily_pnl[(df.index >= 20060101) & (df.index <= 20061231)]
daily_pnl_test
```

```python
sharpe_ratio = np.sqrt(252) * np.mean(daily_pnl_test) / np.std(daily_pnl_test)
sharpe_ratio
```

```python
# model trading costs
one_way_cost = 0.0005
# | w_i - w_(i-1) | is used to calculate cost
traded_weight = weights - np.array(pd.DataFrame(weights).shift())
trading_cost = np.nansum(abs(traded_weight), axis=1) * one_way_cost
daily_pnl_with_txn_cost = daily_pnl - trading_cost
```

```python
daily_pnl_with_txn_cost
```

```python
daily_pnl_test_with_txn_cost = daily_pnl_with_txn_cost[(df.index >= 20060101) & (df.index <= 20061231)]
daily_pnl_test_with_txn_cost
```

```python
sharpe_ratio_with_txn_cost = np.sqrt(252) * np.mean(daily_pnl_test_with_txn_cost) / np.std(daily_pnl_test_with_txn_cost)
sharpe_ratio_with_txn_cost
```
